```python
import pandas as pd
from sklearn.impute import KNNImputer

# Load dataset
df = pd.read_csv('NO2_ugm-3_Delhi_2017_2024.csv')

# Check for datetime or non-numeric columns
print("Column Data Types:\n", df.dtypes)

# Convert datetime column to numeric (if present)
if 'date_column' in df.columns:  # Replace 'date_column' with your actual column name
    df['date_column'] = pd.to_datetime(df['date_column'])
    df['date_numeric'] = df['date_column'].astype(int) // 10**9  # Convert to seconds
    df.drop(columns=['date_column'], inplace=True)  # Drop original datetime column

# Select only numeric columns for imputation
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
print("Numeric Columns for Imputation:", numeric_columns)

# Apply KNN Imputer to numeric columns
imputer = KNNImputer(n_neighbors=5)
df_numeric = pd.DataFrame(imputer.fit_transform(df[numeric_columns]), columns=numeric_columns)

# Recombine with non-numeric columns
non_numeric_columns = df.select_dtypes(exclude=['float64', 'int64']).columns
df_combined = pd.concat([df_numeric, df[non_numeric_columns].reset_index(drop=True)], axis=1)

print("Imputation completed successfully!")
```
